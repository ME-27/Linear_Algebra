__5*.__ Написать на Python функцию для перемножения двух произвольных матриц, не используя NumPy.

```python
def matrixes_product(A, B):
    
    # проверяем на возможность произведения
    if len(A) != len(B[0]):
        print('матрицы нельзя перемножить')
        return None
    
    # транспонируем матрицу B - так удобнее потом перемножать элементы двух матриц
    B_trans = []
    for i in range(len(B[0])):
        new_line = []
        for ii in range(len(B)):
            new_line.append(B[ii][i])
        B_trans.append(new_line)
    
    # создаём болванку под итоговую матрицу
    m_product = [[0] * len(A) for _ in range(len(A))]
    
    # заполняем болванку
    for i in range(len(m_product)):
        for ii in range(len(m_product[i])):
            m_product[i][ii] = [element for element in zip([m for m in A[i]], [n for n in B_trans[ii]])]
            print(m_product[i][ii])  # демонстрация механизма рассчёта каждого элемента итоговой матрицы
            m_product[i][ii] = sum([element[0] * element[1] for element in m_product[i][ii]])
            print(m_product[i][ii])  # демонстрация механизма рассчёта каждого элемента итоговой матрицы
        
    return m_product
```
