__1.__ Найти собственные векторы и собственные значения для линейного оператора, заданного матрицей

### $$A=\begin{pmatrix}-1 & -6\\ 2 & 6\end{pmatrix}$$

найдем собственные значения линейного оператора, составив и решив характеристическое уравнение:

### $$\begin{vmatrix}-1-\lambda & -6\\ 2 & 6-\lambda\end{vmatrix} = 0$$

### $$(-1-\lambda)(6-\lambda)-(-6)\cdot2=0$$

### $$\lambda^{2}-5\lambda+6=0$$

решив квадратное уравнение получаем два собственных значения линейного оператора:

### $$\lambda_{1}=3, \lambda_{2}=2$$

теперь найдем собственные векторы вида $\begin{pmatrix}x_{1}\\ x_{2}\end{pmatrix}$, подставив полученные собственные значения в выражение $\textbf{A}x=\lambda x$:

### $$\begin{pmatrix}a & b\\ c & d\end{pmatrix}\begin{pmatrix}x_{1}\\ x_{2}\end{pmatrix}=\lambda\begin{pmatrix}x_{1}\\ x_{2}\end{pmatrix}.$$

получим

### $$\begin{cases}ax_{1}+bx_{2}=\lambda x_{1}, \\cx_{1}+dx_{2}=\lambda x_{2}.\end{cases}$$ 

для $\lambda_{1}=3$:

### $$\begin{cases}-1x_{1}+(-6)x_{2}=3x_{1}, \\2x_{1}+6x_{2}=3x_{2}\end{cases}$$ 

### $$\begin{cases}x_{1}=-1.5x_{2}, \\0=0\end{cases}$$ 

т.е. при любом $x_{2}$, $x_{1}=-1.5x_{2}$ или:

### $$\begin{cases}x_{1}=-1.5x_{2}, \\x_{2} \in \mathbb{R}\end{cases}$$ 

проверим правильность ответа:

### $$\begin{pmatrix}-1 & -6\\ 2 & 6\end{pmatrix}\begin{pmatrix}-1.5x_{2}\\ x_{2}\end{pmatrix}=$$

### $$=\begin{pmatrix}(-1)(-1.5x_{2})+(-6)x_{2}\\ 2(-1.5x_{2})+6x_{2}\end{pmatrix}=\begin{pmatrix}-4.5x_{2}\\ 3x_{2}\end{pmatrix}=3\begin{pmatrix}-1.5x_{2}\\ x_{2}\end{pmatrix}$$

в случае с $\lambda_{2}=2$ получаем:

### $$\begin{cases}x_{1}=-2x_{2}, \\x_{2} \in \mathbb{R}\end{cases}$$ 

проверим правильность ответа:

### $$\begin{pmatrix}-1 & -6\\ 2 & 6\end{pmatrix}\begin{pmatrix}-2x_{2}\\ x_{2}\end{pmatrix}=$$

### $$=\begin{pmatrix}(-1)(-2x_{2})+(-6)x_{2}\\ 2(-2x_{2})+6x_{2}\end{pmatrix}=\begin{pmatrix}-4x_{2}\\ 2x_{2}\end{pmatrix}=2\begin{pmatrix}-2x_{2}\\ x_{2}\end{pmatrix}$$

__3.__ Пусть линейный оператор задан матрицей

### $$A=\begin{pmatrix}1 & 1\\ -1 & 3\end{pmatrix}$$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

предположим, что вектор $x$ является собственным вектором заданного линейного оператора, тогда должно существовать некоторое вещественное число $\lambda$, при котором 

### $$\begin{pmatrix}
1 & 1\\ -1 & 3\end{pmatrix}\begin{pmatrix}1\\ 1 \end{pmatrix}=\lambda\begin{pmatrix}1\\ 1\end{pmatrix}$$

тогда:

### $$\begin{cases}1\cdot1+1\cdot1=1\cdot\lambda \\ (-1)\cdot1+3\cdot1=1\cdot\lambda\end{cases}\Rightarrow\begin{cases}\lambda=2\\ \lambda=2\end{cases}\Rightarrow\lambda=2$$

раз есть единственное значение $\lambda$, тогда вектор $x$ является собственным вектором линейного оператора $A$
