__1.__ Найти с помощью NumPy SVD для матрицы

### $$\begin{pmatrix}1 & 2 & 0\\ 0 & 0 & 5\\ 3 & -4 & 2\\ 1 & 6 & 5\\ 0 & 1 & 0\end{pmatrix}$$

```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)
```

```python
A = np.array([[1, 2, 0],
              [0, 0, 5],
              [3, -4, 2],
              [1, 6, 5], 
              [0, 1, 0]])
print(f'Матрица A:\n{A}')
```

```python
U, s, W = np.linalg.svd(A)
```
