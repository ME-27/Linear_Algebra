__2.__ –ù–∞–π—Ç–∏ –Ω–æ—Ä–º—ã –≤–µ–∫—Ç–æ—Ä–æ–≤ $(4, 2, 4)$ –∏ $(12, 3, 4)$ –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏.

–Ω–∞–π–¥—ë–º $\ell_{2}$ –Ω–æ—Ä–º—É (–µ–≤–∫–ª–∏–¥–æ–≤—É –Ω–æ—Ä–º—É):

### $$\left\|x\right\| = \sqrt{\sum_{i}|x_{i}|^{2}}$$

–ø—É—Å—Ç—å (4,2,4) = x, –∞ (12,3,4) = y, —Ç–æ–≥–¥–∞ –Ω–æ—Ä–º–∞–º–∏ –≤–µ–∫—Ç–æ—Ä–æ–≤ –±—É–¥—É—Ç:

### $$\left\|x\right\| = \sqrt{4^{2}+2^{2}+4^{2}} = \sqrt{36} = 6$$

–Ω–∞–π–¥—ë–º –∫–æ—Å–∏–Ω—É—Å —É–≥–ª–∞ –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

### $$cos\varphi = \frac{(x,y)}{\left \|x\right \|\cdot\left \|y\right \|}$$

–≥–¥–µ

### $$(x,y) = 4 \cdot 12 + 2 \cdot 3 + 4 \cdot 4 = 70$$

—Ç–æ–≥–¥–∞

### $$cos\varphi = \frac{70}{78} \approx 0.90 $$

–ø—Ä–æ–≤–µ—Ä–∏–º –Ω–∞–π–¥–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ—Å–∏–Ω—É—Å–∞ —Å –ø–æ–º–æ–∑—â—å—é Python, –∞ —Ç–∞–∫–∂–µ –Ω–∞–π–¥—ë–º –∑–Ω–∞—á–µ–Ω–∏–µ —É–≥–ª–∞ –∏–∑ –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—Å–∏–Ω—É—Å–∞:

```python
import numpy as np
from numpy.linalg import norm
```

```python
x = np.array([4, 2, 4])
y = np.array([12, 3, 4])
```

```python
cos_phi = np.dot(x, y) / norm(x) / norm(y)
print(f'–ö–æ—Å–∏–Ω—É—Å —É–≥–ª–∞ –º–µ–∂–¥—É x –∏ y: {cos_phi:.2f}')
print(f'–£–≥–æ–ª –º–µ–∂–¥—É x –∏ y: {np.arccos(cos_phi):.2f}')
```
