### а)

### $$\begin{cases}3x_{1}-x_{2}+x_{3}=4, \\2x_{1}-5x_{2}-3x_{3}=-17, \\x_{1}+x_{2}-x_{3}=0;\end{cases}$$

напишем для решения задачи вот такую замечательную функцию:

```python
import numpy as np
```

```python
def check_compatibility(A_extended):
    
    # из расширенной матрицы создаём обычную
    A = [_[:-1] for _ in A_extended]
    
    # считаем количество неизвестных в системе уравнений
    n = len(A[0])
    
    # переводим матрицы из списков в массивы
    A = np.array(A)
    A_extended = np.array(A_extended)
    
    # считаем ранги
    rank_A = np.linalg.matrix_rank(A)
    rank_A_extended = np.linalg.matrix_rank(A_extended)
    
    print(f'\nматрица A: \n{A}')
    print(f'\nрасширенная матрица A: \n{A_extended}')
    
    print(f'\nранг матрицы A: {rank_A}')
    print(f'ранг расширенной матрицы A: {rank_A_extended}')
    print(f'количество переменных: {n}')
    
    print('\n\nответ:')
    
    # проверяем на совместимость и считаем количество решений
    if rank_A < rank_A_extended:
        print('система несовместна')
    
    elif rank_A == rank_A_extended == n:
        print('система совместна, определена, имеет единственное решение')
        
    elif rank_A == rank_A_extended < n:
        print('система совместна, имеет бесконечное количество решений')
        
    else:
        print('где-то тут ошибка')
    
    return None
```

```python
x = [3, -1, 1, 4]
y = [2, -5, -3, -17]
z = [1, 1, -1, 0]

check_compatibility([x, y, z])
```

__2.__ Найти $L$-матрицу $LU$-разложения для матрицы коэффициентов:

### а)

### $$\begin{pmatrix}1 & 2 & 4 \\ 2 & 9 & 12 \\ 3 & 26 & 30\end{pmatrix}$$

напишем для решения задачи вот такую замечательную функцию:

```python
def form_L_and_U_matrixes(U):
    # функция делает LU разложение некой матрицы
    # функция принимает на вход матрицу
    # функция возвращает матрицы L, U
    
    # порядок/размер матрицы U (L)
    n = len(U)
    # болванка под матрицу L
    L = np.eye(n)
    
    # создание матриц U и L
    for i in range(n):
        for ii in range(i, n):
            try:
                L[ii+1, i] = U[ii+1, i] / U[i, i]
                U[ii+1] = U[ii+1] - U[i] * L[ii+1, i]
            except:
                None
                
    return L, U
```

__3.__ Решить систему линейных уравнений методом $LU$-разложения

### $$\begin{cases}2x_{1}+x_{2}+3x_{3}=1 \\11x_{1}+7x_{2}+5x_{3}=-6 \\9x_{1}+8x_{2}+4x_{3}=-5\end{cases}$$

напишем для решения задачи вот такую замечательную функцию:

```python
def LU_decomposition(A_extended):
    # функция решает систему уравнений путём LU разложения
    # функция принимает на вход систему уравнений в виде расширенной матрицы - A_extended
    # функция возвращает вектор значений x
    
    # вектор значений b
    b = A_extended[:, -1]
    # болванка под матрицу U
    U = A_extended[:, :-1]
    # порядок/размер матрицы U (L)
    n = len(U)
    # болванка под матрицу L
    L = np.eye(n)
    
    # создание матриц U и L
    for i in range(n):
        for ii in range(i, n):
            try:
                L[ii+1, i] = U[ii+1, i] / U[i, i]
                U[ii+1] = U[ii+1] - U[i] * L[ii+1, i]
            except:
                None
                
    def solve_triange_system(triangle_matrix, triangle_matrix_extension):
        # функция решает систему уравнений - матрицу треугольного вида типа L
        # функция принимает на вход саму матрицу - triangle_matrix,
        # а также вектор b из правой стороны уравнения - triangle_matrix_extension
        # функция возвращает вектор значений y из уравнения Ly=b
        
        # также функция решает систему уравнений - матрицу треугольного вида типа U, 
        # если её (матрицу U), а также вектор значений y из уравнения Ly=b, подать на вход развёрнутыми
        # в таком случае функция возвращает вектор значений x из уравнения Ux=y
        # вектор значений x полученный из функции надо развернуть обратно

        result = np.array([0 for _ in range(len(b))], dtype=float)
        for i in range(n):
            line_result = triangle_matrix_extension[i]
            for ii in range(i+1):
                if ii == i:
                    result[i] = line_result / triangle_matrix[i, ii]
                    continue
                line_result -= triangle_matrix[i, ii] * result[ii]

        return result
    
    # расчёт значений y и x, возврат ответа - x
    return solve_triange_system(triangle_matrix=U[::-1, ::-1], 
                     triangle_matrix_extension=solve_triange_system(triangle_matrix=L, triangle_matrix_extension=b)[::-1])[::-1]
```
